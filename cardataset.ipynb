```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
url="C:/Users/LENOVO/Downloads/automobile/imports-85.data"
data=pd.read_csv(url,names=headers)
data.head(10)
```

```python
data.replace("?",np.nan,inplace=True) #replace "?" to NaN
data.head() #show the all rows
```

```python
missed_data=data.isnull() #True for missing data
#also use .notnull() for not missing data
missed_data.head() #True for missing data
```

```python
for column in missed_data.columns.values.tolist(): #count the missing data in each column
    print(column)
    print (missed_data[column].value_counts()) #count the missing data in each column
    print("")    
```

```python
avg_norm_losses = data["normalized-losses"].astype("float").mean(axis=0) #calculate the average of the column
print("Average of normalized-losses:", avg_norm_losses) #print the average of the column
```

```python
data["normalized-losses"].replace(np.nan, avg_norm_losses, inplace=True) #replace NaN by the average
data.head() #show the all rows
```

```python
avg_bore = data['bore'].astype('float').mean(axis=0) #calculate the average of the 
print("Average of bore:", avg_bore) #print the average of the column
```

```python
data["bore"].replace(np.nan, avg_bore, inplace=True) #replace NaN by the average
data.head() #show the all rows
```

```python
avg_stroke = data['stroke'].astype('float').mean(axis=0) #calculate the average of the column
print("Average of stroke:", avg_stroke) #print the average of the column
```

```python
data["stroke"].replace(np.nan, avg_stroke, inplace=True) #replace NaN by the average
data.head() #show the all rows
```

```python
avg_horsepower = data['horsepower'].astype('float').mean(axis=0) #calculate the average of the column
print("Average horsepower:", avg_horsepower) #print the average of the column

data['horsepower'].replace(np.nan, avg_horsepower, inplace=True) #replace NaN by the average

avg_peakrpm = data['peak-rpm'].astype('float').mean(axis=0) #calculate the average of the column
print("Average peak rpm:", avg_peakrpm) #print the average of the column

data['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True) #replace NaN by the average
```

```python
data['num-of-doors'].value_counts() #count the number of doors
```

```python
data['num-of-doors'].value_counts().idxmax() #find the most common type of doors
```

```python
data["num-of-doors"].replace(np.nan, "four", inplace=True) #replace the missing 'num-of-doors' by the most common type of doors
```

```python
data.dropna(subset=["price"], axis=0, inplace=True) #drop the missing values of the column 'price', axis=0 means drop the entire row
data.reset_index(drop=True, inplace=True) #reset the index and drop the old index
```

```python
data.head(25) #show the all rows
```

```python
data.dtypes #check the data type of each column
```

```python
data[["bore", "stroke"]] = data[["bore", "stroke"]].astype("float")
data[["normalized-losses"]] = data[["normalized-losses"]].astype("int")
data[["price"]] = data[["price"]].astype("float")
data[["peak-rpm"]] = data[["peak-rpm"]].astype("float")
```

```python
data.dtypes #check the data type of each column
```

```python
data['city-L/100km'] = 235/data['city-mpg'] #convert mpg to L/100km by mathematical operation (235 divided by mpg)
data.head() #show the all rows
#this is standardization
```

```python
data['highway-L/100km'] = 235/data['highway-mpg'] #convert mpg to L/100km by mathematical operation (235 divided by mpg)
data.head() #show the all rows
#this is standardization
```

```python
data['length'] = data['length']/data['length'].max() #normalize the column 'length'
data['width'] = data['width']/data['width'].max() #normalize the column 'width'
data['height'] = data['height']/data['height'].max() #normalize the column 'height'
data[["length","width","height"]].head() #show the all rows
```

```python
data['horsepower'] = data['horsepower'].astype(int, copy=True) #convert data type to integer
%matplotlib inline
from matplotlib import pyplot
plt.hist(data["horsepower"])
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

```python
bins = np.linspace(min(data["horsepower"]), max(data["horsepower"]), 4) #create 3 bins of equal size bandwidth
bins
#linspace(start_value, end_value, numbers_generated) function returns an array with evenly spaced numbers over a specified interval.
```

```python
group_names = ['Low', 'Medium', 'High']
```

```python
data['horsepower-binned'] = pd.cut(data['horsepower'], bins, labels=group_names, include_lowest=True )
data[['horsepower','horsepower-binned']].head(20)
```

```python
data["horsepower-binned"].value_counts() #count the number of vehicles in each bin
```

```python
%matplotlib inline
from matplotlib import pyplot
pyplot.bar(group_names, data["horsepower-binned"].value_counts())

plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```
