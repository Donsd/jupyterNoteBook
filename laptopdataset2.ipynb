```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
url='C:/Users/LENOVO/jupyterNoteBook/laptop_pricing_dataset_mod1.csv'
data=pd.read_csv(url,header=0)
data.head(10)
```

```python
data.info()
```

```python
data['Screen_Size_cm']=np.round(data[['Screen_Size_cm']],2)
data.head()
```

```python
data.replace('?',np.nan,inplace=True)
missed_data=data.isnull()
missed_data.head(10)
```

```python
for column in missed_data.columns.values.tolist():
    print(column)
    print(missed_data[column].value_counts())
    print("")
```

```python
avg_weight_kg=data['Weight_kg'].astype("float").mean(axis=0)
print('Average of weight-kg:',avg_weight_kg)
```

```python
data['Weight_kg'].replace(np.nan,avg_weight_kg,inplace=True)
data.head()
```

```python
data['Screen_Size_cm'].value_counts()
```

```python
max_Screen_Size=data['Screen_Size_cm'].value_counts().idxmax()
print(max_Screen_Size)
```

```python
data['Screen_Size_cm'].replace(np.nan,max_Screen_Size,inplace=True)
```

```python
data.dtypes
```

```python
data[['Screen_Size_cm']]=data[['Screen_Size_cm']].astype('float')
data[['Weight_kg']]=data[['Weight_kg']].astype('float')
```

```python
data['Screen_Size_cm']=data['Screen_Size_cm']/2.54
data['Weight_kg']=data['Weight_kg']*2.205
data.rename(columns={'Weight_kg':'Weight_pounds'}, inplace=True)

data.head()
```

```python
data.rename(columns={'Screen_Size_cm':'Screen_Size_inch'},inplace=True)
data.head()
```

```python
data['CPU_frequency']=data['CPU_frequency']/data['CPU_frequency'].max()
data.head()
```

```python
bins = np.linspace(min(data['Price']),max(data['Price']),4)
bins
```

```python
Price_bins=['Low','Medium','High']
data['Price-binned']=pd.cut(data['Price'],bins,labels=Price_bins,include_lowest=True)
data[['Price','Price-binned']].tail()
```

```python
data['Price-binned'].value_counts()
```

```python
%matplotlib inline
from matplotlib import pyplot
pyplot.bar(Price_bins,data['Price-binned'].value_counts())

plt.xlabel('Price')
plt.ylabel('count')
plt.ylabel('Price bins')
```

```python
dummy_variable_1=pd.get_dummies(data['Screen'])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'Full HD':'Full-HD-Screen','IPS Panel':'IPS-Panel-Screen'},inplace=True)
dummy_variable_1.head()
```

```python
data=pd.concat([data,dummy_variable_1],axis=1)
data.drop('Screen',axis=1,inplace=True)
data.head()
```
